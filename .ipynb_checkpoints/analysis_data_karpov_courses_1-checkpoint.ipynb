## This notebook is mini project by working with Pandas 


```python
# task 1 - Upload dataset and basis of analysis data 
import pandas as pd

bookings = pd.read_csv('data/bookings.csv', sep=';')
print(bookings.shape)
print(bookings.dtypes)
bookings_head = bookings.head(7)
```

```python
# task 2 - print name columns in lower registry and replace space on chart lower _

new_name_columns = [ncolumn.replace(' ', '_').lower() for ncolumn in list(bookings.columns)]
bookings.rename(columns=dict(zip(bookings.columns, new_name_columns)), inplace=True)
```

```python
bookings.columns
```

```python
# task 3 - find country by users who to commit success action in the booking hotel
(bookings
 .query('is_canceled == 0')
 .groupby(['country'])
 .agg({'is_canceled': 'count'})
 .sort_values('is_canceled', ascending=False)).head()
```

```python
# task 4 - calculate mean count of the night by different tipes
(bookings
.groupby('hotel')
.agg({'stays_total_nights': 'mean'}))
```

```python
# task 5
(bookings
.query('assigned_room_type != reserved_room_type'))
```

```python
#task 6 - analysis Arrival full date
# point 1 - booking in 2016
(bookings
.query('arrival_date_year == 2016')
 ['arrival_date_month'].value_counts().idxmax())
```

```python
# point 1 - booking in 2017
(bookings
.query('arrival_date_year == 2017')
 ['arrival_date_month'].value_counts())
```

```python
# point 3 - check hotel with type City Hotel
check_city_hotel = (bookings
                    .query('is_canceled == 1 & hotel == "City Hotel"')
                    .groupby(['arrival_date_year', 'arrival_date_month'])
                    .agg({'is_canceled': 'count'})
                    .sort_values(by=['arrival_date_year', 'is_canceled'], ascending=False))

check_city_hotel
```
